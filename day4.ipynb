```python
import sys
import os

current_dir = os.getcwd()
parent_dir = os.path.dirname(current_dir)
if parent_dir not in sys.path:
    sys.path.insert(0, parent_dir)

from get_data import get_data_from_url
url = 'https://adventofcode.com/2024/day/4/input'

raw_data = get_data_from_url(url)
```

```python
# day 1: 
raw_data_lines = raw_data.splitlines()
horizontal_sum = 0
for line in raw_data_lines:
    horizontal_sum += line.count('XMAS')
    horizontal_sum += line.count('SAMX')
horizontal_sum


vertical_sum = 0
for i in range (0, len(raw_data_lines)-3):
    for k in range (len(raw_data_lines[0])):
        if (raw_data_lines[i][k]=="X" and \
            raw_data_lines[i+1][k]=='M' and \
            raw_data_lines[i+2][k]=='A' and \
            raw_data_lines[i+3][k]=='S') or \
            (raw_data_lines[i][k]=="S" and \
            raw_data_lines[i+1][k]=='A' and \
            raw_data_lines[i+2][k]=='M' and \
            raw_data_lines[i+3][k]=='X'):
            vertical_sum+=1

vertical_sum + horizontal_sum 

raw_data_lines = raw_data.splitlines()
horizontal_sum = 0
for line in raw_data_lines:
    horizontal_sum += line.count('XMAS')
    horizontal_sum += line.count('SAMX')
horizontal_sum


pos_diagonal_sum = 0
for i in range (0, len(raw_data_lines)-3):
    for k in range (0, len(raw_data_lines[0])-3):
        if (raw_data_lines[i][k]=="X" and \
            raw_data_lines[i+1][k+1]=='M' and \
            raw_data_lines[i+2][k+2]=='A' and \
            raw_data_lines[i+3][k+3]=='S') or \
            (raw_data_lines[i][k]=="S" and \
            raw_data_lines[i+1][k+1]=='A' and \
            raw_data_lines[i+2][k+2]=='M' and \
            raw_data_lines[i+3][k+3]=='X'):
            pos_diagonal_sum+=1

neg_diagonal_sum = 0
for i in range (0, len(raw_data_lines)-3):
    for k in range (3,len(raw_data_lines[0])):
        if (raw_data_lines[i][k]=="X" and \
            raw_data_lines[i+1][k-1]=='M' and \
            raw_data_lines[i+2][k-2]=='A' and \
            raw_data_lines[i+3][k-3]=='S') or \
            (raw_data_lines[i][k]=="S" and \
            raw_data_lines[i+1][k-1]=='A' and \
            raw_data_lines[i+2][k-2]=='M' and \
            raw_data_lines[i+3][k-3]=='X'):
            neg_diagonal_sum+=1



vertical_sum + horizontal_sum + pos_diagonal_sum + neg_diagonal_sum     
```

```python
def calc_ans(raw_data_lines=raw_data_lines):
    vertical_sum = 0
    for i in range (0, len(raw_data_lines)-2):
        for k in range (len(raw_data_lines[0])-2):
            if (raw_data_lines[i][k]=='M' and raw_data_lines[i][k+2]=='M' and\
                raw_data_lines[i+1][k+1]=='A' and \
                raw_data_lines[i+2][k]=='S' and raw_data_lines[i+2][k+2]=='S') \
                or \
                (raw_data_lines[i][k]=='S' and raw_data_lines[i][k+2]=='M' and\
                raw_data_lines[i+1][k+1]=='A' and \
                raw_data_lines[i+2][k]=='S' and raw_data_lines[i+2][k+2]=='M') \
                or \
                (raw_data_lines[i][k]=='M' and raw_data_lines[i][k+2]=='S' and\
                raw_data_lines[i+1][k+1]=='A' and \
                raw_data_lines[i+2][k]=='M' and raw_data_lines[i+2][k+2]=='S') \
                or \
                (raw_data_lines[i][k]=='S' and raw_data_lines[i][k+2]=='S' and\
                raw_data_lines[i+1][k+1]=='A' and \
                raw_data_lines[i+2][k]=='M' and raw_data_lines[i+2][k+2]=='M'):
                vertical_sum+=1
    return vertical_sum

calc_ans()
```
